## 4) Create Derived Columns

```python
# Order Year
# Purpose: Extracts order year for yearly sales analysis.
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Order Year'] = df['Order Date'].dt.year
```

```python
# Order Month
# Purpose: Extracts order month for monthly sales trend analysis.
df['Order Month'] = df['Order Date'].dt.month_name()
print(df['Order Month'])
```

```python
# Profit Margin
#Purpose: Calculates percentage profit earned from sales.
#Formula:Profit Margin = (Profit / Sales) Ã 100
df['Profit Margin'] = (df['Profit'] / df['Sales']) * 100
```

```python
# Delivery Days
#Purpose: Calculates delivery duration between order date and shipping date.
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

## Visualization 1 : Sales Distribution

```python
# VISUALIZATION 1 : SALES DISTRIBUTION (HISTOGRAMS)
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30, kde=True)  #bins=30 divides the sales values into 30 intervals, kde=True adds a smooth density curve to show the distribution pattern.
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')  # Representing the number of records in each sales range.
plt.show()
```

## Visualization 13 : Category vs Region Sales Analysis (Heatmap)

Objective

To analyze sales performance across different categories and regions and identify high-performing and low-performing combinations.

Variables Used:
- Category
- Region
- Sales

```python
# Visualization 13 : Category vs Region Sales Analysis (Heatmap)
# Create Pivot Table
pivot_sales = pd.pivot_table(
    df,
    values='Sales',
    index='Category',
    columns='Region',
    aggfunc='sum'
)
# Visualization
plt.figure(figsize=(10,6))
sns.heatmap(pivot_sales,
            annot=True,  # Displays sales values inside each cell.
            fmt='.0f',   # Remove decimals
            cmap='YlGnBu')
plt.title('Category vs Region Sales')
plt.show()
```


A Heatmap cannot directly analyze raw data efficiently when you want to compare Category vs Region Sales.
The pivot table summarizes the data into a matrix format that the heatmap can easily visualize.

How to Read the Heatmap
- Rows â Categories
- Columns â Regions
- Cell Values â Total Sales

cmap='YlGnBu'
- Applies Yellow-Green-Blue color scheme.
- Darker colors represent higher sales.
- Lighter colors represent lower sales.

A pivot table is created to summarize sales data by category and region, transforming raw records into a matrix format that enables effective visualization and comparison through a heatmap.

## Step 5 - Correlation Analysis
      
Correlation Analysis measures the strength and direction of the relationship between two or more numerical variables.

Purpose:
- To determine whether variables are related.
- To measure how strongly they are related.
- To identify positive, negative, or no relationship between variables.
- To support decision-making by understanding variable dependencies.
  
Correlation Values

- +1            - Perfect Positive Correlation
- +0.5 to +0.9	- Strong Positive Correlation
- 0	            - No Correlation
- -0.5 to -0.9	- Strong Negative Correlation
- -1	        - Perfect Negative Correlation

## Visualization 15 : Correlation Matrix (Heatmap)

Objective:

To measure and visualize the strength and direction of relationships between numerical variables such as Sales, Profit, Quantity, and Discount.

Variables Used:

- Sales
- Profit
- Quantity
- Discount

```python
# Visualization 15: Correlation Matrix (Heatmap)
corr_matrix = df[['Sales','Profit',
                  'Quantity','Discount']].corr() # Select Numerical Columns and Calculate Correlation
plt.figure(figsize=(8,5))
sns.heatmap(corr_matrix,
            annot=True,
           cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Uses a color scale:

- Red = Positive correlation
- Blue = Negative correlation
- Light colors = Weak correlation

The correlation analysis indicates that Sales and Profit have a moderate positive relationship (0.48), suggesting that higher sales generally lead to higher profits. Discount shows a negative correlation with Profit (-0.22), indicating that increasing discounts may reduce profitability. Quantity has only weak relationships with Sales and Profit, while Discount and Quantity exhibit almost no correlation. This suggests that discount policies have a greater impact on profit than on sales volume
